```python
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as mim
from tensorflow.keras.layers.experimental import preprocessing
```

```python
def list_dir(name_):
  import os
  for dir_path, dirnames, filenames in os.walk(name_):
    print(f'There are {len(dirnames)} directories and {len(filenames)} images in {dir_path}')
```

```python
def plot_loss_curve(history):
  """
  Returns seperate losss curves for training and validation metrics
  Args:
  history: Tensorflow history object
  Returns:
  plots of training and Validation loss and accuracy
  """
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  acc = history.history['acc']
  val_acc = history.history['val_acc']

  epochs = range(len(history.history['loss']))
  plt.figure()
  plt.plot(epochs, loss, label='Training_loss')
  plt.plot(epochs, val_loss, label='val_loss')
  plt.title(label='Loss')
  plt.xlabel(xlabel='epochs')
  plt.legend()
  plt.figure()
  plt.plot(epochs, acc, label='Training_Accuracy')
  plt.plot(epochs, val_acc, label='val_acc')
  plt.title(label='Accuracy')
  plt.xlabel(xlabel='epochs')
  plt.legend()
```

```python
train_dir = '/content/drive/MyDrive/Cotton Disease/train'
test_dir = '/content/drive/MyDrive/Cotton Disease/test'
val_dir = '/content/drive/MyDrive/Cotton Disease/val'
```

```python
IMG_SIZE = (224, 224)
BATCH = 32
```

```python
# Loading data from the path

# Training Data
train_data = tf.keras.preprocessing.image_dataset_from_directory(
    directory = train_dir,
    label_mode ='categorical',
    image_size = IMG_SIZE,
    batch_size = BATCH)

# Testing Data
test_data = tf.keras.preprocessing.image_dataset_from_directory(
    directory = test_dir,
    label_mode ='categorical',
    image_size = IMG_SIZE,
    batch_size = BATCH)

# Validation Data
val_data = tf.keras.preprocessing.image_dataset_from_directory(
    directory = val_dir,
    label_mode ='categorical',
    image_size = IMG_SIZE,
    batch_size = BATCH)
```

```python
class_names = train_data.class_names
```

```python
class_names
```

## Building Model and Preprocessing image files

```python
input_shape = (224, 224, 3)
batch = 32
```

```python

data_augmentation = tf.keras.Sequential([
    preprocessing.RandomZoom(0.2),
    preprocessing.RandomHeight(0.2),
    preprocessing.RandomWidth(0.2),
    preprocessing.RandomFlip()], name='data_augmentation')
```

```python
len(class_names)
```

```python
base_model = tf.keras.applications.EfficientNetB0(include_top=False) # back bone of our model
base_model.trainable = False
```

```python
inputs = tf.keras.Input(shape=input_shape, name='input_shape')
augmented_layer = data_augmentation(inputs, training=True)
getting_op_from_base = base_model(augmented_layer, training=False)
# getting_op_from_base = base_model(inputs, training=False) # Training is False because we dont want to impact loaded weights and batch Normalization
pooling_op = tf.keras.layers.GlobalAveragePooling2D(name='Pooling_layer')(getting_op_from_base)
outputs = tf.keras.layers.Dense(4, activation='softmax', name='output_layer')(pooling_op)
model_1 = tf.keras.Model(inputs, outputs)
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
model_path = r"/content/drive/MyDrive/Cotton Disease/models/model_vgg16.h5"
callbacks = [
        ModelCheckpoint(model_path, verbose=1, save_best_only=True),
    ]
```

```python
model_1.compile(optimizer=tf.keras.optimizers.Adam(),
                loss=tf.keras.losses.CategoricalCrossentropy(),
                metrics=['acc'])
```

```python
model_1.summary()
```

```python
model_hist = model_1.fit(train_data,
            epochs=20,
            validation_data=val_data,
            callbacks = callbacks
            )
```

```python
plot_loss_curve(model_hist)
```

```python
model_1.evaluate(test_data)
```
